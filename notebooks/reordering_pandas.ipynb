```python
import pandas as pd
df = pd.DataFrame({'buyer': ('A', 'A', 'B', 'A','B','A'),
                   'seller' : ('B','B','A','B','A','B'),
                  'Type': ('1', '1', '2', '1','2','1')})

df
```

```python
df['rank'] = df.groupby(['Type','seller','buyer']).cumcount()
df.sort_values(['rank'])
```
